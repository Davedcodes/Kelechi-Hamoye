```python
df = df.drop(columns = 'stab')
```

```python
x = df.drop(['stabf'], axis = 1)
y = df.stabf
```

```python
x
```

```python
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size =0.2, random_state = 1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_transform = scaler.fit_transform(x_train)
x_train_transform = pd.DataFrame(x_train_transform, columns = x_train.columns)
```

```python
x_train_transform.head()
```

```python
x_test_transform = scaler.transform(x_test)
x_test_transform = pd.DataFrame(x_test_transform, columns = x_test.columns)
```

```python
x_test_transform.head()
```

```python
TP = 355
TN = 120
FP = 1480
FN = 45
precision = TP/(TP + FP)
recall = TP/(TP + FN)
f1_score = 2*((precision* recall)/(precision+recall))

round(f1_score,4)
```

```python
TP2 = 82
TN2 = 98
FP2 = 2
FN2 = 18
precision_2 = TP2/(TP2 + FP2)
recall_2 = TP2/(TP2 + FN2)
f1_score = 2*((precision_2 * recall_2)/(precision_2 + recall_2))
print('recall: %f' % recall_2)
```

```python
from sklearn.ensemble import RandomForestClassifier
random_class = RandomForestClassifier(random_state=1)
random_class.fit(x_train_transform,y_train)
```

```python
predict = random_class.predict(x_test_transform)
```

```python
from sklearn.metrics import recall_score, accuracy_score, precision_score, f1_score, confusion_matrix
print("Accuracy score {}".format(round(accuracy_score(y_test, predict), 4)))
print("Precision score for label stable %.3f" % (precision_score(y_test, predict, pos_label='stable')))
print("Recall score for label stable {}".format(round(recall_score(y_test, predict, pos_label='stable'), 4)))
print("F1 score %.3f" % (f1_score(y_test, predict, pos_label='stable')))
```
